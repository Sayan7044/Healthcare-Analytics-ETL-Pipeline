```python
import numpy as np
import pandas as pd
```

```python
#Loading data
df = pd.read_csv('messy_healthcare_dataset.csv')
df_cpy1 = df.copy()
df_cpy1
```

```python
df_cpy1.info()
```

```python
df_cpy1.isnull().sum()
```

```python
#Remove records with Paitent ID = Null
df_cpy1.dropna(subset=['Patient_ID'])
```

```python
# Convert Age to numeric
df_cpy1['Age'] = pd.to_numeric(df_cpy1['Age'],errors='coerce')
```

```python
df_cpy1['Age'].dtype
```

```python
# Fill the NaN values in age with the median value
df_cpy1['Age'] = df_cpy1['Age'].fillna(df_cpy1['Age'].median())
df_cpy1
```

```python
#Wait_Time(min) -> numeric
df_cpy1['Wait_Time(min)'] = pd.to_numeric(df_cpy1['Wait_Time(min)'],errors='coerce')
df_cpy1
```

```python
df_cpy1['Wait_Time(min)'].dtype
```

```python
# fill the NaN with the mean waiting tim value
df_cpy1['Wait_Time(min)'] = df_cpy1['Wait_Time(min)'].fillna(round(df_cpy1['Wait_Time(min)'].mean(),2))
df_cpy1
```

```python
# Name Cleaning
df_cpy1['Name'] = df_cpy1['Name'].str.strip().str.title()
df_cpy1
```

```python
#Gender cleaning
df_cpy1['Gender'] = df_cpy1['Gender'].str.strip().str.lower()
df_cpy1
```

```python
df_cpy1['Gender'] = df_cpy1['Gender'].replace({
    'm':'Male',
    'male':'Male',
    'f':'Female',
    'female':'Female',
    'Maleale':'Male',
    'FemaleeMaleale':'Male'
})

df_cpy1
```

```python
#Department cleaning
df_cpy1['Department'].unique()
```

```python
df_cpy1['Department'] = df_cpy1['Department'].str.strip().str.title()
df_cpy1
```

```python
df_cpy1['Department'] = df_cpy1['Department'].replace({
    'Neuro':'Neurology'
})

df_cpy1
```

```python
#City cleaning
df_cpy1['City'].unique()
```

```python
df_cpy1['City'] = df_cpy1['City'].str.strip().str.title()
df_cpy1
```

```python
df_cpy2 = df_cpy1.copy()
df_cpy2
```

```python
df_cpy2['Appointment_Date'] = pd.to_datetime(df_cpy2['Appointment_Date'],format='mixed',errors='coerce')
df_cpy2
```

```python
df_cpy2['Year'] = df_cpy2['Appointment_Date'].dt.year
df_cpy2['Month'] = df_cpy2['Appointment_Date'].dt.month
df_cpy2['Day'] = df_cpy2['Appointment_Date'].dt.day
df_cpy2['Weekday'] = df_cpy2['Appointment_Date'].dt.day_name()

df_cpy2
```

```python
df_cpy2['Bill_Amount'].dtype
```

```python
df_cpy2['Bill_Amount'] = pd.to_numeric(df_cpy2['Bill_Amount'],errors='coerce')
df_cpy2
```

```python
df_cpy2['Bill_Amount'] = df_cpy2['Bill_Amount'].fillna(df_cpy2['Bill_Amount'].median())
df_cpy2
```

```python
df_cpy2 = df_cpy2.drop_duplicates()
df_cpy2
```

```python
df_cpy2['Payment_Status'] = df_cpy2['Payment_Status'].str.strip().str.title()
df_cpy2
```

## Feature Engineering ##

```python
#Age Group
df_cpy3 = df_cpy2.copy()
df_cpy3
```

```python
# Age Group
df_cpy3['Age_Group'] = pd.cut(df_cpy3['Age'],bins=[0,18,35,60,100],labels=['Child','Young','Adult','Senior'])
df_cpy3
```

```python
#High bill
df_cpy3['High_Bill']=df_cpy3['Bill_Amount']>5000
df_cpy3
```

```python
# long waiting
df_cpy3['Long_Wait'] = df_cpy3['Wait_Time(min)'] > 60
df_cpy3
```
